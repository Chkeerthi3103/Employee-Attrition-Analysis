```python
data=pd.read_csv('HREmployeeAttrition.csv')
data=data.drop_duplicates()
print('Number of instances = %d\nNumber of attributes = %d' % (data.shape[0], data.shape[1]))
print('------------------------------')
data.isnull().sum() 
```

```python
#MonthlyIncome,YearsAtCompany,Department,TotalWorkingYears for all the employees whose age is greater than 25 and Attrition is yes
data.filter(items=["MonthlyIncome","YearsAtCompany","Department","TotalWorkingYears"]).loc[(data['Age']>=25) & (data['Attrition'] == "Yes")] 
```

```python
## Histogram
data[['Age','Attrition','BusinessTravel','DailyRate','Department','DistanceFromHome','Education','EducationField',
      'EmployeeCount','EmployeeNumber','EnvironmentSatisfaction','Gender','HourlyRate','JobInvolvement','JobLevel',
      'JobRole','JobSatisfaction','MaritalStatus' ,'MonthlyIncome','MonthlyRate','NumCompaniesWorked','Over18',
      'OverTime' ,'PercentSalaryHike','PerformanceRating','RelationshipSatisfaction','StandardHours',
      'StockOptionLevel','TotalWorkingYears','TrainingTimesLastYear','WorkLifeBalance','YearsAtCompany',
      'YearsInCurrentRole',
      'YearsSinceLastPromotion','YearsWithCurrManager']].hist(figsize=(15, 15), bins=20, alpha = 0.75, color = 'red')
plt.tight_layout()
plt.show()
```

```python
### Checking for outliers

data.boxplot(figsize=(20,8))
plt.subplots_adjust(bottom=0.25)
plt.xticks(rotation=90)
```

```python
##Heat map
import seaborn as sns
colunas = ['Age','DailyRate','DistanceFromHome','Education','EnvironmentSatisfaction','HourlyRate','JobInvolvement',
           'JobLevel', 'JobSatisfaction','MonthlyIncome','MonthlyRate','NumCompaniesWorked',
           'PercentSalaryHike','PerformanceRating','RelationshipSatisfaction','StockOptionLevel','TotalWorkingYears',
           'TrainingTimesLastYear',
           'WorkLifeBalance','YearsAtCompany','YearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager']
plt.figure(figsize=(18,9))
sns.heatmap(data[colunas].corr(), annot=True, fmt='.1f')
```

```python
#count plot
colunas = ['WorkLifeBalance','TrainingTimesLastYear','StockOptionLevel',
    'RelationshipSatisfaction','PerformanceRating','NumCompaniesWorked',
   'JobInvolvement', 'JobLevel', 'JobSatisfaction',
   'EnvironmentSatisfaction','Education']

plt.figure(figsize=(12,36))
for i,col in enumerate(colunas):
    axes = plt.subplot(13,2, i + 1)
    sns.countplot(x=data[col], hue=data['Department'], palette=['#65F085','#8565f0'])
plt.tight_layout()
plt.show()
```

```python
## histogram plot gender
hist = ['Age', 'DailyRate','DistanceFromHome','HourlyRate','MonthlyIncome',
       'MonthlyRate']

plt.figure(figsize=(13,25))
for i,col in enumerate(hist):
    axes = plt.subplot(6,2, i + 1)
    sns.histplot(x=data[col], hue=data['Attrition'])
plt.tight_layout()
plt.show()
```

```python
X=data.select_dtypes(include=[object])
X
plt.figure(figsize=(18,60))
for i,col in enumerate(X):
    axes = plt.subplot(13,2, i + 1)
    sns.countplot(x=data[col], hue=data['Gender'])
plt.tight_layout()
plt.show()
```
